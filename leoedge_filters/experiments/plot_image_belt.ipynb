```python
import folium
import os
import re
import xml.etree.ElementTree as ET
import pickle
import pandas as pd
```

```python
all_metadata=open('../data/all_metadata','r').readlines()
# randomly choose 1000 images
```

```python
metadata=[]
for filename in all_metadata:
    filename=filename.strip()
    if filename.endswith('.xml'):
        # Extract the image id from the filename using regular expressions
        match = re.search(r'(\w+)_3B_AnalyticMS_metadata.xml', filename)
        # if match:
        image_id = match.group(1)
        satellite_id = image_id[-4:]
        tree = ET.parse(filename)
        root = tree.getroot()
        # Extract the four corners of the image
        coords = root.find(
            ".//gml:coordinates",
            namespaces={'gml': 'http://www.opengis.net/gml'})
        # Convert the string of coordinates to a list of floats
        coords = [[float(x) for x in coord.split(',')] for coord in coords.text.split()]
        day=filename.split('/')[-1].split('_')[0]
        # Add the image id, center latitude, and center longitude to the metadata list
        metadata.append({
            'id': image_id,
            'points': coords,
            'satellite_id': satellite_id,
            'day': day
        })
```

```python
merged_data=pd.DataFrame(metadata)
merged_data
```
