```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
import datetime
```

```python
passes = ['Overlap 1', 'Overlap 2', 'Overlap 3', 'Overlap 4', 'No Overlap 1', 'No Overlap 2']
valid = [5, 1, 15, 6]
corrupt = [3, 1, 4, 7]
start1 = 1676080925
end1 = 1676081410
start2 = 1676166028
end2 = 1676166581
start3 = 1676212969
end3 = 1676213495
start4 = 1676251143
end4 = 1676251729
total = [5/(np.floor((end1-start1)/3)),1/(np.floor((end2-start2)/3)),15/(np.floor((end3-start3)/3)),6/(np.floor((end4-start4)/3)),14/60,13/60]
barWidth = 0.25 # the width of the bars
br1 = np.arange(len(passes))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
c = ['r', 'r','r','r','g','g']
# Make the plot
plt.bar(passes, total, color = c, width = barWidth,
        edgecolor ='grey')

#plt.xticks([r + barWidth for r in range(len(passes))],passes)
plt.tick_params(axis='both', which='major',labelsize=18)
#plt.legend(fontsize=18, loc='upper left')
plt.xlabel(' ',fontsize=18)
plt.xticks(rotation=45)
plt.ylabel('Packet Reception Rate',fontsize=18)
plt.figure(figsize=(10,6))
```

```python
fig, ax = plt.subplots()
width = 0.35       # the width of the bars: can also be len(x) sequence
ax.bar(passes, valid, width, label='Valid')
ax.bar(passes, corrupt, width, bottom=valid,
       label='Corrupt')
ax.bar(passes, np.array(total)-np.array(corrupt)-np.array(valid), width, bottom=np.array(valid)+np.array(corrupt),label='Undetected')
plt.tick_params(axis='both', which='major',labelsize=18)
plt.legend(fontsize=18, loc='upper right')
plt.xlabel(' ',fontsize=18)
plt.ylabel('# of Packets',fontsize=18)
plt.figure(figsize=(10,6))
```
